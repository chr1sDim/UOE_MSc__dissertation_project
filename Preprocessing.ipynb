## Interpolation and hour intervals with few measurements

```python
#insert number of house to be loaded
house_nr=1
data_raw = data_provider.load_house(house_nr)
data = pd.DataFrame()
data['Aggregate'] = data_raw.Aggregate.resample('min').mean()
```

```python
data['Hour'] = pd.DatetimeIndex(data.index).hour
data['Date'] = pd.DatetimeIndex(data.index).date
```

```python
def null_r(inp):
    count=0
    for i in inp:
        if np.isnan(i):
            count += 1
    return count

nulls = data['Aggregate'].groupby([data['Date'],data['Hour']]).apply(null_r)
```

```python
nulls = nulls.reset_index()
```

```python
data['Nulls'] = nulls.Aggregate.values
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.Nulls , color='b')
plt.xticks(rotation=45)
plt.title('House nr. 1 Unregistered Electricity Consumption')
plt.xlabel('Time')
plt.ylabel('Consumption (Kwh)')
plt.grid(True)
#plt.savefig('figures/house'+str(house_nr)+'_unregistered.pdf')
plt.show()
```
